```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib
import joblib 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from weather_data_processing import preprocess_data
import numpy as np

# Завантаження і попередня обробка даних
%load_ext autoreload
%autoreload 2
from weather_data_processing import preprocess_data
raw_df = pd.read_csv('weatherAUS.csv')
data = preprocess_data(raw_df)

X_train = data['train_X']
train_targets = data['train_y']
X_val = data['val_X']
val_targets = data['val_y']
X_test = data['test_X']
test_targets = data['test_y']

# Модель дерева рішень з обрізанням (ccp_alpha)
model = DecisionTreeClassifier(random_state=42, max_leaf_nodes=128)
model.fit(X_train, train_targets)

# Перевірка наявності каталогу 'models' і створення його, якщо не існує
# if not os.path.exists('models'):
#    os.makedirs('models')

# Збереження моделі за допомогою joblib
joblib.dump(model, "models/aussie_rain_.joblib")
print("Модель успішно збережена!")

# Оцінка на валідаційних даних
val_accuracy = model.score(X_val, val_targets)
print(f"Точність на валідаційній вибірці: {val_accuracy:.2f}")

# Прогнозування на тестовому наборі
test_preds = model.predict(X_test)
test_accuracy = accuracy_score(test_targets, test_preds)
print(f"Точність на тестовій вибірці: {test_accuracy:.2f}")
```

```python
# Прогнозування для одного з екземплярів
input_array = X_train.iloc[0].values  # Вибір першого екземпляра
prediction = model.predict(np.expand_dims(input_array, axis=0))  # Прогноз для одного екземпляра
print(f"Прогноз для першого екземпляра: {prediction}")
```

```python
import joblib
from sklearn.ensemble import RandomForestClassifier

# Завантаження даних
data = joblib.load('aussie_rain.joblib')

# Перевірка типу
print("Loaded data type:", type(data))

# Якщо це словник, вивести ключі
if isinstance(data, dict):
    print("Keys in the dictionary:", data.keys())
    
    # Вивести інформацію про кожен об'єкт у словнику
    for key, value in data.items():
        print(f"Key: {key}, Type: {type(value)}")
        
        # Якщо це об'єкт моделі, можна вивести її атрибути
        if isinstance(value, RandomForestClassifier):
            # Виводимо лише основні атрибути
            print(f"RandomForestClassifier parameters for {key}:")
            print(f"  n_estimators: {value.n_estimators}")
            print(f"  max_depth: {value.max_depth}")
            print(f"  min_samples_split: {value.min_samples_split}")
            print(f"  min_samples_leaf: {value.min_samples_leaf}")
            print(f"  max_features: {value.max_features}")
        elif hasattr(value, 'get_params'):
            # У разі, якщо це інший об'єкт моделі
            print(f"Params for {key}: {value.get_params()}")
        else:
            print(f"Value for {key}: {value}")
else:
    print("The loaded data is not a dictionary.")
```
